Importing libraries and dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds = pd.read_csv('Position_Salaries.csv')
x = ds.iloc[:, 1:-1].values
y = ds.iloc[:, -1].values
print(x)
print(y)
```

Training the linear regression on the dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
```

Training the polynomial regression on the dataset

```python
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree = 4)
x_poly = pr.fit_transform(x)
lr2 = LinearRegression()
lr2.fit(x_poly, y)
```

Visualising linear regression results

```python
plt.scatter(x, y, color = 'orange')
plt.plot(x, lr.predict(x), color = 'green')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Visualising polynomial regression results

```python
plt.scatter(x, y, color = 'orange')
plt.plot(x, lr2.predict(pr.fit_transform(x)), color = 'green')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Visualising polynomial regression results for smooth curve

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'orange')
plt.plot(x_grid, lr2.predict(pr.fit_transform(x_grid)), color = 'green')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Predicting new result with linear regression

```python
lr.predict([[5.5]])
```

Predicting new result with polynomial regression

```python
lr2.predict(pr.fit_transform([[5.5]]))
```
